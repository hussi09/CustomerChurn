```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
telco = pd.read_csv('Churn.csv')
print(telco.info())
```

```python
# how many churners does the dataset have, and how many non-churners
print(telco['Churn'].value_counts())
```

```python
# Group telco by 'Churn' and compute the mean, std,
df = telco[[ 'Churn','CustServ_Calls', 'Vmail_Message']]
print(df.groupby(['Churn']).mean())
print(df.groupby(['Churn']).std())
```

```python
# Based on the abov results, Churners make more customer service calls than non-churners.
```

```python
# Count the number of churners and non-churners by State
print(telco.groupby('State')['Churn'].value_counts())
```

```python
# Exploring feature distribution
import seaborn as sns
sns.set()
# Visualize the distribution of 'Day_Mins','Eve_Mins','Night_Mins','Intl_Mins'.
_ = sns.distplot(telco['Day_Mins'])
plt.show()
```

```python
# Customer service calls and churn
sns.boxplot(x = 'Churn',
            y = 'CustServ_Calls',
            data = telco)
plt.show()
```

```python
# Remove outliers
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x = 'Churn',
            y = 'CustServ_Calls',
            data = telco, sym ="")
plt.show()
```

```python
#Add a third variable to this plot - 'Vmail_Plan'
#to visualize whether or not having a voice mail plan affects the number of customer service calls or churn.
sns.boxplot(x = 'Churn',
            y = 'CustServ_Calls',
            data = telco,
            sym = "",
            hue = "Vmail_Plan")

# Display the plot
plt.show()
```

```python
sns.boxplot(x = 'Churn',
            y = 'CustServ_Calls',
            data = telco,
            sym = "",
            hue = "Intl_Plan")
# Add "Intl_Plan" as a third variable
# Display the plot
plt.show()
```
